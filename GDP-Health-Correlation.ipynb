```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("all_data.csv")
```

```python
print(df.head())
```

```python
print(df['Country'].unique())
```

```python
#The Max, Min, mean, and median life expectancy of each country.
```

```python
countries = df['Country'].unique()

for country in countries:
    stats = df[df['Country'] == country]['Life expectancy at birth (years)']
    print(f"{country} - Max: {np.max(stats)}, Min: {np.min(stats)}, Mean: {np.mean(stats)}, Median: {np.median(stats)}")
```

```python
#The Max, Min, mean, and median GDP of each country.
```

```python
countries = df['Country'].unique()

for country in countries:
    stats = df[df['Country'] == country]['GDP']
    print(f"{country} - Max: {np.max(stats)}, Min: {np.min(stats)}, Mean: {np.mean(stats)}, Median: {np.median(stats)}")
```

```python
# plot GDP and Life Expectancy to see the general trends.
```

```python
plt.scatter(df["GDP"], df["Life expectancy at birth (years)"], color="blue")
sns.regplot(x=df["GDP"], y=df["Life expectancy at birth (years)"], scatter_kws={'color':'blue'}, line_kws={'color':'red'})
plt.annotate("Highest GDP - USA", xy=(65000, 79), xytext=(60000, 85),
             arrowprops=dict(facecolor='red', shrink=0.05))
plt.annotate("Lowest GDP - Zimbabwe", xy=(2000, 65), xytext=(5000, 70),
             arrowprops=dict(facecolor='green', shrink=0.05))
plt.xlabel("GDP (in billion $)")
plt.ylabel("Life Expectancy (years)")
plt.title("GDP vs Life Expectancy - Trend Line Added")
plt.show()
```

```python
# plot the spread of life expectancy across the six countries
```

```python
sns.boxplot(x=df['Country'], y=df['Life expectancy at birth (years)'])
sns.set_palette("coolwarm")
plt.xticks(rotation=45)
plt.xlabel('Country', fontsize=14)
plt.ylabel('Life Expectancy (years)', fontsize=14)
plt.title('Life Expectancy vs GDP', fontsize=16, fontweight='bold')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

##  Conclusion
- The analysis shows a **strong correlation** between GDP and life expectancy.
- Countries with higher GDP tend to have **better healthcare & longer life expectancy**.
- **China's rapid growth** after economic reforms played a major role in improving health statistics.
- **Zimbabwe's lower GDP** indicates challenges in life expectancy improvements.
